```python
threshold=len(df)*0.6
df=df.dropna(axis=1,thresh=threshold)
```

A threshold is set at 60% of the total rows, and dropna(axis=1, thresh=threshold) removes columns that do not meet this minimum number of non-null values.

```python
len(df.columns)
```

 The len(df.columns) function then shows the updated number of columns remaining after cleaning.

```python
currency_cols=['GDP','Health Exp/Capita','Tourism Inbound','Tourism Outbound']

for col in currency_cols:
  df[col]=df[col].str.replace(',','').str.replace('$','').astype(float)
```

First, selected currency-related columns (like GDP and Health Expenditure per Capita) are converted from string format to numeric by removing commas and dollar signs, and then casting them to float.

```python
numeric_cols=[]
categorical_cols=[]
for col in df.columns:
  if df[col].dtype=='object':
    categorical_cols.append(col)
  else:
    numeric_cols.append(col)
print('Numeric Columns:',numeric_cols)
print('Categorical Columns:',categorical_cols)
```

 Next, the dataset columns are separated into numerical and categorical features based on their data types, which helps in applying appropriate analysis and preprocessing techniques.

```python
df.info()
```

Now columns are of desired data type

```python
for col in numeric_cols:
  df[col]=df[col].fillna(df[col].mean())
```

It iterates through each numeric column and replaces null values with the mean of that column using fillna(). This ensures that missing data is imputed without significantly affecting the overall distribution of the dataset.
